```python
# Another way to generate the series is
mylynxts_aw = pd.Series(mylynx_df['trappings'].values, index = pd.date_range('31/12/1821', periods = 114, freq = 'A-DEC'))
```

```python
mylynxts_aw
```

```python
from statsmodels.tsa.stattools import adfuller
def stationarity_test(timeseries):
    print("Dickey-Fuller test results:")
    df_test = adfuller(timeseries, autolag = 'AIC')
    df_output = pd.Series(df_test[0:4], index = ["Test Statistic", "p-value", "#Lags Used", "Num. of observations Used"])
    print(df_output)
```

```python
stationarity_test(mylynxts_aw)
```

 The ADFuller test removes autocorrelation from the series and as a low p_value was returned (<0.05 ) this inidcates the series is stationary.

Autocorrelation is an important characteristic to understand as it indicates the relation of an observation and its previous one. For ARIMA models it is importer to specify the order.

```python
def plot_rolling(timeseries, window):
    rol_mean = timeseries.rolling(window).mean()
    rol_std = timeseries.rolling(window).std()
    
    fig = plt.figure(figsize = (12, 8))
    og = plt.plot(timeseries, color = "blue", label = "Original")
    mean = plt.plot(rol_mean, color = "red", label = "Rolling Mean")
    std = plt.plot(rol_std, color = "black", label = "Rolling Std")
    plt.legend(loc = "best")
    plt.title("Rolling Mean and Standard Deviation (window =" + str(window) +")")
    plt.show()
```

```python
plot_rolling(mylynxts_aw,10)
```

```python
#EWMA

def plot_ewma(timeseries, alpha):
    expw_ma = timeseries.ewm(alpha=alpha).mean()
    
    fig = plt.figure(figsize = (12, 8))
    og_line = plt.plot(timeseries, color = "blue", label = "Original")
    exwm_line = plt.plot(expw_ma, color = "red", label = "EWMA")
    plt.legend(loc = "best")
    plt.title("EWMA (alpha= "+str(alpha)+")")
    plt.show()
```

```python
plot_ewma(mylynxts_aw, 0.5)
```

## Arima model

```python
from statsmodels.tsa.arima_model import ARIMA
from matplotlib.pyplot import rcParams
#p: AR (PACF)
#D: differenctiate 
#q: MA ( ACF)
```

```python
rcParams['figure.figsize'] = 15, 6
```

```python
# date format for Arima models
mylynxts = pd.Series(mylynx_df['trappings'].values, index = pd.DatetimeIndex(data=(tuple(pd.date_range('31/12/1821', periods=114,freq='A-DEC'))), freq='A-DEC'))
```

```python
#Look bakc at the ACF nad PACF graphs we put the order for ARIMA as (2,0,0)
# Second we try with (3,0,0)
model = ARIMA(mylynxts, order=(4,0,0))
results_AR = model.fit()
plt.plot(mylynxts)
plt.plot(results_AR.fittedvalues, color = 'RED')
```

```python
results_AR.summary()
```

### All patterns should be captured by the model, only randomness should stay in the residuals. The are the containers of randomness ( what happens by chance or can't be explaines through maths).

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(211)
fig = plot_acf(results_AR.resid, lags=20, ax=ax1)
```

```python
# import fucntion for norm dist
from scipy.stats import norm
```

```python
plt.figure(figsize =(10, 6))
plt.hist(results_AR.resid, bins = 'auto', density = True, rwidth = 0.85, label = 'Residuals')
mu, std = norm.fit(results_AR.resid)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'm', linewidth =2)
plt.grid(axis='y', alpha = 0.2)
plt.xlabel('Residuals')
plt.ylabel('Density')

plt.title('Residuals 2,0,0 vs Normal Distribution - Mean = '+str(round(mu,2))+', Std = '+str(round(std,2)))
plt.show()
```
